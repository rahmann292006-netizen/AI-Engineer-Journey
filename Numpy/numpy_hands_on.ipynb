```python
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)

y = np.sin(x)

plt.plot(x,y)

plt.title("Sine Wave")

plt.xlabel("X")

plt.ylabel("sin(x)")

plt.grid(True)

plt.show()
```
