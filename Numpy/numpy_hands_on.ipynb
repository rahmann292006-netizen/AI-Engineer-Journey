```python
# ==============================
# Day 3 - NumPy Indexing & Slicing
# ==============================

import numpy as np

# Create a 3x4 array
arr = np.arange(1,13).reshape(3,4)

print("Original Array:")
print(arr)

# ------------------------------
# Indexing
# ------------------------------

print("\nFirst Row:")
print(arr[0])

print("\nSecond Row:")
print(arr[1])

print("\nFirst Element:")
print(arr[0,0])

print("\nLast Element:")
print(arr[-1,-1])

# ------------------------------
# Slicing
# ------------------------------

print("\nFirst Two Columns:")
print(arr[:,0:2])

print("\nLast Two Rows:")
print(arr[1:,:])

print("\nMiddle Elements:")
print(arr[:,1:3])

# ------------------------------
# Boolean Masking
# ------------------------------

print("\nElements Greater Than 6:")
print(arr[arr > 6])

print("\nEven Numbers:")
print(arr[arr % 2 == 0])

# ------------------------------
# Fancy Indexing
# ------------------------------

print("\nFirst and Third Row:")
print(arr[[0,2]])

print("\nColumns 1 and 3:")
print(arr[:,[0,2]])

# ------------------------------
# Iteration
# ------------------------------

print("\nIterating through Array:")

for row in arr:
    print(row)

# ------------------------------
# Mini Challenge
# ------------------------------

marks = np.array([35,78,91,44,67,82,29,99,56,73])

print("\nStudent Marks:")
print(marks)

print("Highest Marks:", np.max(marks))
print("Lowest Marks:", np.min(marks))
print("Average Marks:", np.mean(marks))

print("Marks Above 75:")
print(marks[marks > 75])

print("Marks Below 40:")
print(marks[marks < 40])

print("Sorted Marks:")
print(np.sort(marks))
```

```python
# -----------------------------
# Broadcasting with Arrays
# -----------------------------

a = np.array([[1], [2], [3]])
b = np.array([10, 20, 30])

print("\nBroadcasting Result:")
print(a + b)
```

```python
# -----------------------------
# Vectorized Operations
# -----------------------------

numbers = np.arange(1, 11)

print("\nNumbers:")
print(numbers)

print("\nSquare:")
print(numbers ** 2)

print("\nCube:")
print(numbers ** 3)

print("\nSquare Root:")
print(np.sqrt(numbers))
```

```python
# -----------------------------
# Conditional Operations
# -----------------------------

print("\nNumbers Greater Than 5:")
print(numbers[numbers > 5])

print("\nEven Numbers:")
print(numbers[numbers % 2 == 0])

# -----------------------------
# Universal Functions (ufuncs)
# -----------------------------

angles = np.array([0, np.pi/2, np.pi])

print("\nSin:")
print(np.sin(angles))

print("\nCos:")
print(np.cos(angles))

print("\nLog:")
print(np.log(np.array([1, 10, 100])))

print("\nExponential:")
print(np.exp(np.array([1, 2, 3])))
```

```python
# ==========================================
# Day 5 - Stacking, Splitting, Sorting & Searching
# ==========================================

import numpy as np

# ---------------------------------
# Vertical Stack
# ---------------------------------

a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])

print("Vertical Stack")
print(np.vstack((a, b)))

# ---------------------------------
```

```python
# ---------------------------------
# Horizontal Stack
# ---------------------------------

print("\nHorizontal Stack")
print(np.hstack((a, b)))
```

```python
# ---------------------------------
# Splitting Arrays
# ---------------------------------

arr = np.arange(1, 13)

print("\nSplit into 3 parts")
print(np.split(arr, 3))
```

```python
# -----------------------------
# Matrix Addition
# -----------------------------

print("\nMatrix Addition:")
print(A + B)

# -----------------------------
# Matrix Subtraction
# -----------------------------

print("\nMatrix Subtraction:")
print(A - B)

# -----------------------------
# Matrix Multiplication
# -----------------------------

print("\nMatrix Multiplication:")
print(np.dot(A, B))
```

```python

# -----------------------------
# Transpose
# -----------------------------

print("\nTranspose of A:")
print(A.T)

# -----------------------------
# Determinant
# -----------------------------

print("\nDeterminant of A:")
print(np.linalg.det(A))

# -----------------------------
# Inverse
# -----------------------------

C = np.array([[4, 7],
              [2, 6]])

print("\nMatrix C:")
print(C)

print("\nInverse of C:")
print(np.linalg.inv(C))
```

```python
# -----------------------------
# Rank
# -----------------------------

print("\nRank of C:")
print(np.linalg.matrix_rank(C))

# -----------------------------
# Trace
# -----------------------------

print("\nTrace of C:")
print(np.trace(C))

# -----------------------------
# Eigenvalues & Eigenvectors
# -----------------------------

values, vectors = np.linalg.eig(C)

print("\nEigenvalues:")
print(values)

print("\nEigenvectors:")
print(vectors)
```
