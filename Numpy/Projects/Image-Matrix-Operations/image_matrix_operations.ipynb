```python
img = Image.open("sample.jpg")
```

```python
img_array = np.array(img)
```

```python
print("Shape :", img_array.shape)
print("Size :", img_array.size)
print("Dimensions :", img_array.ndim)
print("Data Type :", img_array.dtype)
```

```python
horizontal = np.fliplr(img_array)

plt.imshow(horizontal)
plt.title("Horizontal Flip")
plt.axis("off")
plt.show()
```

```python
vertical = np.flipud(img_array)

plt.imshow(vertical)
plt.title("Vertical Flip")
plt.axis("off")
plt.show()
```

```python
rotate90 = np.rot90(img_array)

plt.imshow(rotate90)
plt.title("Rotate 90°")
plt.axis("off")
plt.show()
```

```python
crop = img_array[30:120, 30:120]

plt.imshow(crop)
plt.title("Cropped Image")
plt.axis("off")
plt.show()
```

```python
height, width = img_array.shape[:2]

crop = img_array[
    height//4 : 3*height//4,
    width//4 : 3*width//4
]

plt.imshow(crop)
plt.title("Center Crop")
plt.axis("off")
plt.show()
```

```python
bright = np.clip(img_array + 50,0,255).astype(np.uint8)

plt.imshow(bright)
plt.title("Brightness Increased")
plt.axis("off")
plt.show()
```

```python
dark = np.clip(img_array - 50,0,255).astype(np.uint8)

plt.imshow(dark)
plt.title("Brightness Decreased")
plt.axis("off")
plt.show()
```

```python
gray = np.mean(img_array,axis=2)

plt.imshow(gray,cmap="gray")
plt.title("Grayscale")
plt.axis("off")
plt.show()
```

```python
negative = 255 - img_array

plt.imshow(negative)
plt.title("Negative Image")
plt.axis("off")
plt.show()
```
