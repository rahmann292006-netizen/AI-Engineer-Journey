```python
# ==========================================
# Day 7 - Pandas Fundamentals
# ==========================================

import pandas as pd

# -----------------------------
# Creating a Series
# -----------------------------

s = pd.Series([10, 20, 30, 40, 50])

print("Series:")
print(s)

# -----------------------------
# Creating a DataFrame
# -----------------------------

data = {
    "Name": ["Rahman", "Ali", "Sara"],
    "Age": [20, 21, 19],
    "City": ["Raichur", "Bangalore", "Hyderabad"]
}

df = pd.DataFrame(data)

print("\nDataFrame:")
print(df)

# -----------------------------
# Basic Information
# -----------------------------

print("\nShape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nInfo:")
print(df.info())

print("\nDescription:")
print(df.describe())

# -----------------------------
# Head & Tail
# -----------------------------

print("\nHead:")
print(df.head())

print("\nTail:")
print(df.tail())
```

```python
# ==========================================
# Day 8 - Pandas Indexing, Selection & Filtering
# ==========================================

import pandas as pd

# -----------------------------
# Create a DataFrame
# -----------------------------

data = {
    "Name": ["Rahman", "Ali", "Sara", "Ahmed", "Fatima"],
    "Age": [20, 22, 19, 23, 21],
    "Marks": [85, 78, 92, 88, 95],
    "City": ["Raichur", "Bangalore", "Hyderabad", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n")
print(df)

# -----------------------------
# Selecting Columns
# -----------------------------

print("\nSingle Column:")
print(df["Name"])

print("\nMultiple Columns:")
print(df[["Name", "Marks"]])

# -----------------------------
# Selecting Rows
# -----------------------------

print("\nFirst Three Rows:")
print(df[0:3])

# -----------------------------
# Using loc[]
# -----------------------------

print("\nUsing loc[]:")
print(df.loc[1])

print("\nRows 1 to 3:")
print(df.loc[1:3])

# -----------------------------
# Using iloc[]
# -----------------------------

print("\nUsing iloc[]:")
print(df.iloc[2])

print("\nRows 0 to 2:")
print(df.iloc[0:3])

# -----------------------------
# Boolean Filtering
# -----------------------------

print("\nStudents with Marks > 85")
print(df[df["Marks"] > 85])

print("\nStudents Age > 20")
print(df[df["Age"] > 20])

# -----------------------------
# Multiple Conditions
# -----------------------------

print("\nMarks > 80 and Age > 20")
print(df[(df["Marks"] > 80) & (df["Age"] > 20)])

# -----------------------------
# Sorting
# -----------------------------

print("\nSort by Marks (Ascending)")
print(df.sort_values("Marks"))

print("\nSort by Marks (Descending)")
print(df.sort_values("Marks", ascending=False))

# -----------------------------
# Reset Index
# -----------------------------

print("\nReset Index")
print(df.reset_index(drop=True))
```
