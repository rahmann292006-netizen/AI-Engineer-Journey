```python
# ==========================================
# Day 7 - Pandas Fundamentals
# ==========================================

import pandas as pd

# -----------------------------
# Creating a Series
# -----------------------------

s = pd.Series([10, 20, 30, 40, 50])

print("Series:")
print(s)

# -----------------------------
# Creating a DataFrame
# -----------------------------

data = {
    "Name": ["Rahman", "Ali", "Sara"],
    "Age": [20, 21, 19],
    "City": ["Raichur", "Bangalore", "Hyderabad"]
}

df = pd.DataFrame(data)

print("\nDataFrame:")
print(df)

# -----------------------------
# Basic Information
# -----------------------------

print("\nShape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nInfo:")
print(df.info())

print("\nDescription:")
print(df.describe())

# -----------------------------
# Head & Tail
# -----------------------------

print("\nHead:")
print(df.head())

print("\nTail:")
print(df.tail())
```

```python
# ==========================================
# Day 8 - Pandas Indexing, Selection & Filtering
# ==========================================

import pandas as pd

# -----------------------------
# Create a DataFrame
# -----------------------------

data = {
    "Name": ["Rahman", "Ali", "Sara", "Ahmed", "Fatima"],
    "Age": [20, 22, 19, 23, 21],
    "Marks": [85, 78, 92, 88, 95],
    "City": ["Raichur", "Bangalore", "Hyderabad", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n")
print(df)

# -----------------------------
# Selecting Columns
# -----------------------------

print("\nSingle Column:")
print(df["Name"])

print("\nMultiple Columns:")
print(df[["Name", "Marks"]])

# -----------------------------
# Selecting Rows
# -----------------------------

print("\nFirst Three Rows:")
print(df[0:3])

# -----------------------------
# Using loc[]
# -----------------------------

print("\nUsing loc[]:")
print(df.loc[1])

print("\nRows 1 to 3:")
print(df.loc[1:3])

# -----------------------------
# Using iloc[]
# -----------------------------

print("\nUsing iloc[]:")
print(df.iloc[2])

print("\nRows 0 to 2:")
print(df.iloc[0:3])

# -----------------------------
# Boolean Filtering
# -----------------------------

print("\nStudents with Marks > 85")
print(df[df["Marks"] > 85])

print("\nStudents Age > 20")
print(df[df["Age"] > 20])

# -----------------------------
# Multiple Conditions
# -----------------------------

print("\nMarks > 80 and Age > 20")
print(df[(df["Marks"] > 80) & (df["Age"] > 20)])

# -----------------------------
# Sorting
# -----------------------------

print("\nSort by Marks (Ascending)")
print(df.sort_values("Marks"))

print("\nSort by Marks (Descending)")
print(df.sort_values("Marks", ascending=False))

# -----------------------------
# Reset Index
# -----------------------------

print("\nReset Index")
print(df.reset_index(drop=True))
```

```python
import pandas as pd
import numpy as np

# -------------------------------------
# Day 9 - Missing Values & Data Cleaning
# -------------------------------------

data = {
    "Name": ["Rahman", "Ali", "Sara", "Ahmed", "Fatima", "Ali"],
    "Age": [20, np.nan, 19, 23, np.nan, np.nan],
    "Marks": [85, 78, np.nan, 88, 95, 78],
    "City": ["Raichur", "Bangalore", "Hyderabad", None, "Mumbai", "Bangalore"]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# Missing Values
print("\nMissing Values")
print(df.isnull())

print("\nMissing Values Count")
print(df.isnull().sum())

# Fill Missing Values
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())
df["City"] = df["City"].fillna("Unknown")

print("\nAfter Filling Missing Values")
print(df)

# Duplicate Rows
print("\nDuplicate Rows")
print(df.duplicated())

df = df.drop_duplicates()

print("\nAfter Removing Duplicates")
print(df)

# Rename Column
df = df.rename(columns={"Marks": "Score"})

print("\nRenamed Columns")
print(df)

# Data Summary
print("\nData Information")
print(df.info())

print("\nStatistical Summary")
print(df.describe())
```

```python
import pandas as pd

# Day 10 - Data Transformation

data = {
    "Name": ["Rahman", "Ali", "Sara", "Ahmed"],
    "Gender": ["M", "M", "F", "M"],
    "Salary": [50000, 60000, 55000, 70000],
    "City": ["BLR", "HYD", "DEL", "BLR"]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# apply()
df["Salary"] = df["Salary"].apply(lambda x: x + 5000)

# map()
df["Gender"] = df["Gender"].map({"M":"Male","F":"Female"})

# replace()
df["City"] = df["City"].replace({
    "BLR":"Bangalore",
    "HYD":"Hyderabad",
    "DEL":"Delhi"
})

# String Operations
df["Upper_Name"] = df["Name"].str.upper()
df["Name_Length"] = df["Name"].str.len()

# New Column
df["Bonus"] = df["Salary"] * 0.10

print("\nUpdated DataFrame")
print(df)
```

```python
import pandas as pd

# Day 11 - GroupBy Basics

data = {
    "Department":["IT","HR","IT","Finance","HR","Finance","IT"],
    "Employee":["Rahman","Ali","Sara","Ahmed","Fatima","John","David"],
    "Salary":[50000,45000,65000,70000,48000,72000,60000]
}

df = pd.DataFrame(data)

print(df)

print("\nGroup Size")
print(df.groupby("Department").size())

print("\nAverage Salary")
print(df.groupby("Department")["Salary"].mean())

print("\nMaximum Salary")
print(df.groupby("Department")["Salary"].max())

print("\nMinimum Salary")
print(df.groupby("Department")["Salary"].min())

print("\nTotal Salary")
print(df.groupby("Department")["Salary"].sum())

print("\nCount")
print(df.groupby("Department")["Salary"].count())
```
